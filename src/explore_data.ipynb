```python
import os
import pandas as pd

# Check if the file exists and try different paths
possible_paths = [
    "../data/simulate_activity.csv",
    "./data/simulate_activity.csv",
    "data/simulate_activity.csv",
    os.path.join(os.path.dirname(os.getcwd()), "data", "simulate_activity.csv")
]

file_path = None
for path in possible_paths:
    if os.path.exists(path):
        file_path = path
        break

if file_path:
    df = pd.read_csv(file_path)
    print(f"File loaded from: {file_path}")
    display(df.head())
    print(f"Shape: {df.shape}")
    df.info()
else:
    print(f"Current working directory: {os.getcwd()}")
    print("File not found. Please check if the file exists at the correct location.")
    print("Looking for: simulate_activity.csv in the data folder")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Data Overview
print("=" * 50)
print("DATA OVERVIEW")
print("=" * 50)
print(f"\nDataset shape: {df.shape}") # ignore 
print(f"\nColumn names and types:\n{df.dtypes}")
print(f"\nMissing values:\n{df.isnull().sum()}")
print(f"\nBasic statistics:\n{df.describe()}")

# Check for duplicates
print(f"\nDuplicate rows: {df.duplicated().sum()}")

# Data Distribution
print("\n" + "=" * 50)
print("DATA DISTRIBUTION")
print("=" * 50)

# Categorical columns analysis
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\n{col} - Unique values: {df[col].nunique()}")
    print(df[col].value_counts().head())

# Numerical columns visualization
numerical_cols = df.select_dtypes(include=[np.number]).columns
fig, axes = plt.subplots(len(numerical_cols), 1, figsize=(10, 4 * len(numerical_cols)))
for idx, col in enumerate(numerical_cols):
    if len(numerical_cols) > 1:
        axes[idx].hist(df[col], bins=30, edgecolor='black')
        axes[idx].set_title(f"Distribution of {col}")
    else:
        axes.hist(df[col], bins=30, edgecolor='black')
        axes.set_title(f"Distribution of {col}")
plt.tight_layout()
plt.show()

# Correlation analysis
if len(numerical_cols) > 1:
    plt.figure(figsize=(10, 8))
    sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', center=0)
    plt.title("Correlation Matrix")
    plt.show()

print("\nExploration complete!")
```
